```python
subset = songs[songs['artist'] != songs['artist_name']]

results = {}

for i,row in subset.iterrows():
    
    artist = row['artist'].replace(" ","+")
    song = row['song'].replace(" ","+")
    query = f"{song}+{artist}"

    search = sp.search(query, type='track')
```

```python
cols = ['album', 'artists', 'available_markets', 'disc_number', 'duration_ms', 'explicit', 'external_ids', 
        'external_urls', 'href', 'id', 'is_local', 'name', 'popularity', 'preview_url', 'track_number', 
        'type', 'uri', 'artist', 'artist_id', 'song_index']

df = pd.DataFrame(columns = cols)
for i,row in test.iterrows():
    main_df = pd.DataFrame(row['tracks']['items'], columns = cols)
    for j,record in main_df.iterrows():
        main_df.loc[j,'artist'] = record['artists'][0]['name']
        main_df.loc[j,'artist_id'] = record['artists'][0]['id']
        main_df.loc[j,'song_index'] = i
    
    df = df.append(main_df, ignore_index=True)
    
    
df
```

```python
idx_to_track_id = {}

for i in idx:
    if float(i) in df['song_index'].to_list():
        filtered = df[df['song_index'] == float(i)]
        for j,row in filtered.iterrows():
            if row['artist'] == subset.loc[i,'artist']:
                idx_to_track_id[i] = [row['id'], row['preview_url'], row['name'], row['artist'], row['artist_id']]
                
idx_to_track_id
```

```python
songs_cp = songs.copy()

for key, value in idx_to_track_id.items():
    songs_cp.loc[key,'track_id'] = value[0]
    songs_cp.loc[key,'preview_url'] = value[1]
    songs_cp.loc[key,'track_name'] = value[2]
    songs_cp.loc[key,'artist_name'] = value[3]
    songs_cp.loc[key,'artist_id'] = value[4]
    
songs_cp
```

```python
subset = songs_cp[songs_cp['artist'] != songs_cp['artist_name']]
subset[:50]
```

```python
artists = subset['artist'].to_list()

results = []
for artist in artists:
    query = artist.replace(" ","+")
    search = sp.search(query, type='artist',market='US')
    results.append(search)
    
results
```
